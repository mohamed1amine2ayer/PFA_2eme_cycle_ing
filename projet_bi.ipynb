```python
cross_tab_YELLOW= pd.crosstab(index=df_22['FTR'], columns=df_22['HR'])
cross_tab_YELLOW
```

```python
cross_tab_RED = pd.crosstab(index=df_22['FTR'], columns=df_22['HR'])
cross_tab_RED
```

```python
import numpy as np
from scipy.stats import chi2_contingency

# Create a contingency table of observed values


# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(cross_tab_YELLOW)

# Print the results
print("Chi-square statistic:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected values:", expected)
```

```python
# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(cross_tab_RED)

# Print the results
print("Chi-square statistic:", chi2)
print("p-value:", p)
print("Degrees of freedom:", dof)
print("Expected values:", expected)
```

the calculated test statistic is less than or equal to the critical value, fail to reject the null hypothesis. it means that there is not a significant association between the two variables.

```python
# Remove multiple columns by name
df_22 =df_22.drop(['HR', 'HY'], axis=1)  # remove the 'HR' and 'HY' columns
```

```python
from scipy.stats import f_oneway

class_a_scores = df_22[df_22['FTR'] == 'A']['HF']
class_d_scores = df_22[df_22['FTR'] == 'D']['HF']
class_h_scores = df_22[df_22['FTR'] == 'H']['HF']

f_stat, pval = f_oneway(class_a_scores, class_d_scores, class_h_scores)

print('F-statistic =', f_stat)
print('P-value =', pval)
```

the p-value is less than 0.05, we can reject the null hypothesis and conclude that there is a significant difference in the means of the exam scores across the three classe.
